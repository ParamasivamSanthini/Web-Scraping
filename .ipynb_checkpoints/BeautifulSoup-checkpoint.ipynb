# Web Scraping##

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
import numpy as np

   
movie_titles = []     
years_released = []
time = []
imdb_ratings = []
genre = []
votes = []
pages = np.arange(1, 1001, 50)  
for page in pages:
    url = "https://www.imdb.com/search/title/?groups=top_1000&start=" + str(page)
    html = requests.get(url)
    soup = BeautifulSoup(html.text, "html.parser")
    movie_div = soup.find_all("div", class_="lister-item mode-advanced")
  
    print(page)
    for movieSection in movie_div:
        name = movieSection.h3.a.text
        movie_titles.append(name)
        year = movieSection.h3.find("span", class_="lister-item-year").text
        years_released.append(year)
        ratings = movieSection.strong.text
        imdb_ratings.append(ratings)
        group = movieSection.find("span", class_="genre").text.strip()
        genre.append(group)
        runTime = movieSection.find("span", class_="runtime").text
        time.append(runTime)
        nv = movieSection.find_all("span", attrs={"name": "nv"})
        vote = nv[0].text
        votes.append(vote)
movies = pd.DataFrame(
    {
        "Movie": movie_titles,
        "Year": years_released,
        "RunTime": time,
        "imdb": imdb_ratings,
        "Genre": genre,
        "votes": votes,
    }
)


print(movies)
movies.to_csv("moviess.csv", index=False, header=True)
```

```python
dataset = pd.read_csv("moviess.csv")
dataset
```

```python
# Data Pre processing
#1.Remove ()in the year
#2.Change min to minutes in runtime
#3.Removing comma from votes


dataset["Year"] = dataset["Year"].str.extract("(\\d+)").astype(int)
dataset["RunTime"] = dataset["RunTime"].str.replace("min", "minutes")
dataset["votes"] = dataset["votes"].str.replace(",", "").astype(int)
dataset
```

```python
#Removing special char and symbols
dataset["Movie"] = dataset["Movie"].str.replace(",", " ")
dataset["Movie"] = dataset["Movie"].str.replace(":", " ")
dataset["Movie"] = dataset["Movie"].str.replace("-", " ")
dataset["Movie"] = dataset["Movie"].str.replace("Â", " ")
dataset["Movie"] = dataset["Movie"].str.replace("Ã»", " ")
dataset["Movie"] = dataset["Movie"].str.replace("!", " ")
dataset
```

```python
dataset.dropna()   #No nan values
```

## Analysis##

```python
# High no of rating movie

from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
counts = dataset["Movie"].value_counts()
counts.index = counts.index.map(str)
wordcloud = WordCloud(background_color="white").generate_from_frequencies(counts)
plt.figure(figsize=[10,10])
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
#Cross checked word cloud with bar chart

import seaborn as sns 

counts = dataset["Movie"].value_counts().nlargest(15)
plt.figure(figsize=(10, 6))
counts.plot(kind='barh')
```

```python
#Plot movie released

import matplotlib.pyplot as plt
plt.xlabel("Year")
dataset['Year'].plot(kind='hist',bins=10)
```
